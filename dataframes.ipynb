# All the datasets loaded from the pydataset library will be pandas dataframes.

```python

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
print(df)
```

1) Copy the code from the lesson to create a dataframe full of student grades.

```python
# a. Create a column named passing_english that indicates 
# whether each student has a passing grade in english.

df['passing_english'] = df.english > 70

df
```

```python
# b. Sort the english grades by the passing_english column. 
# How are duplicates handled? 
# Answer: It;s sorted by the index when there is a duplicate number.

df.sort_values(by='passing_english')
```

```python
# c. Sort the english grades first by passing_english and then 
# by student name. 

# All the students that are failing english should be first, 
# and within the students that are failing english they should 
# be ordered alphabetically. 


# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)

df.sort_values(by=['passing_english', 'name'], ascending = [True, True])
```

```python
# d. Sort the english grades first by passing_english, and then 
# by the actual english grade, similar to how we did in the 
# last step.

df.sort_values(by=['passing_english', 'english'], ascending = [False, False])

```

```python
# e. Calculate each students overall grade and add it as a 
# column on the dataframe. The overall grade is the average of 
# the math, english, and reading grades.

df['Overall_grade'] = (df['math'] + df['english'] + df['reading'])/3


df
```

2) Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:



```python
# data('mpg', show_doc=True) # view the documentation for the dataset
mpg = data('mpg') # load the dataset and store it in a variable
```

```python
# How many rows and columns are there?
#  Answer: 234 rows × 11 columns

mpg.shape
```

```python
# What are the data types of each column?

mpg.dtypes
```

```python
# Summarize the dataframe with .info and .describe

mpg.info()
mpg.describe()
```

```python
# Rename the cty column to city.

mpg = mpg.rename(columns={'cty': 'city'})
```

```python
# Rename the hwy column to highway.

mpg = mpg.rename(columns={'hwy': 'highway'})
```

```python
mpg
```

```python
# Do any cars have better city mileage than highway mileage?
# Answer: NO

mpg.sort_values(by=['city', 'highway'], ascending = [False, False])


```

```python
mpg[(mpg.city > mpg.highway)]
```

```python
# Create a column named mileage_difference 
# this column should contain the difference between 
# highway and city mileage for each car.

mpg['mileage_difference'] = mpg.highway - mpg.city

mpg
```

```python
# Which car (or cars) has the highest mileage difference?

mpg.sort_values(by=['mileage_difference', 'manufacturer', 'model'], ascending = [False, False, False]).nlargest(1, 'mileage_difference', keep = 'all')

```

```python
mpg.sort_values(by=['mileage_difference', 'manufacturer', 'model'], ascending = [False, False, False]).mileage_difference.max()


```

```python
mpg[mpg.mileage_difference == mpg.sort_values(by=['mileage_difference', 'manufacturer', 'model'], ascending = [False, False, False]).mileage_difference.max()]
```

```python
# Which compact class car has the lowest highway mileage? 
# Answer volkswagen jetta 23:highway
# The best?
# Answer: volkswagen jetta 44:highway

compacts = mpg[mpg['class']== 'compact']
```

```python
compacts.sort_values(by=['highway']).nsmallest(1, 'highway', keep = 'all')


```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.

mpg['average_mileage'] = (mpg['highway'] + mpg['city'])/2

```

```python
mpg
```

```python
# Which dodge car has the best average mileage? 
# Answer: dodge caravan 2wd
# The worst? 2xdodge ram 1500, dodge durango 4wd, dodge dakota 4wd

cars = mpg[mpg['manufacturer'] == 'dodge']


```

```python
cars
```

```python
cars.sort_values(by=['average_mileage']).nlargest(1, 'average_mileage', keep = 'all').head()

```
