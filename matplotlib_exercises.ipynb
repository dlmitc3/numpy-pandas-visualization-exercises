1.) Use matplotlib to plot the following equation:

                 y = x^2 - x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
# Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np
import math

# Create the vectors X and Y
x = range(-19, 20)
y = [n ** 2 for n in x]

# Create the plot
plt.scatter(x, y)
plt.title('Testing123')
plt.xlabel('$x^2 - x + 2$')
plt.ylabel('$y = x^2 - x + 2$')

# Show the plot
plt.show()
```

You can use functions from the math module to help implement some of the equations above.

```python

# generate data
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# customize the chart
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');

```

```python
# generate data
x2 = range(0,20)
y2 = [n**3 for n in x2]

#plot data
plt.plot(x2,y2)

#customize the chart
plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
x3 = [x * 0.1 for x in range(-50,100)]
y3 = [2**n for n in x3]

#plot data
plt.plot(x3,y3)

# Create the plot
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')

# Show the plot
plt.show()
```

```python
# specify range of x values
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# plot the data
plt.plot(x4,y4)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

```python
plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(221)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#subplot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/x$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)
plt.annotate("some text", xy = (2,0.1), xytext = (2, 0.6), arrowprops = {});


plt.tight_layout()
```
