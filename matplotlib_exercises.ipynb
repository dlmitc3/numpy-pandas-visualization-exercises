```python
# Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np

# Create the vectors X and Y
x = range(-19, 20)
y = [n ** 2 for n in x]

# Create the plot
plt.scatter(x, y)
plt.title('Testing123')
plt.xlabel('$x^2 - x + 2$')
plt.ylabel('$y = x^2 - x + 2$')

# Show the plot
plt.show()
```
